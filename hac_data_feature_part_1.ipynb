```python
def generate_numeric_time_features(df, datetime_col='CREATE_DATETIME'):
    df = df.copy()
    df[datetime_col] = pd.to_datetime(df[datetime_col])

    df['app_year'] = df[datetime_col].dt.year
    df['app_month'] = df[datetime_col].dt.month
    df['app_day'] = df[datetime_col].dt.day
    df['app_hour'] = df[datetime_col].dt.hour
    df['app_minute'] = df[datetime_col].dt.minute
    df['app_dayofweek'] = df[datetime_col].dt.dayofweek

    # Бинарные признаки строго как 0/1
    df['is_night'] = df['app_hour'].between(0, 6).astype(int)
    df['is_off_hours'] = df['app_hour'].isin([0, 1, 2, 3, 4, 5, 23]).astype(int)
    df['is_start_of_month'] = (df['app_day'] <= 3).astype(int)
    df['is_end_of_month'] = df[datetime_col].dt.is_month_end.astype(int)
    df['is_weekend'] = (df['app_dayofweek'] >= 5).astype(int)

    df['app_seconds_since_midnight'] = (
        df['app_hour'] * 3600 +
        df['app_minute'] * 60 +
        df[datetime_col].dt.second
    )

    return df
```
